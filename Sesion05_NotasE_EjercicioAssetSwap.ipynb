# Ejemplo de una extración (scrapping) sencillo de datos

```python
import requests
from bs4 import BeautifulSoup

# URL de la página
url = "https://en.wikipedia.org/wiki/History_of_Federal_Open_Market_Committee_actions"

# Hacer la solicitud
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

# Encontrar todas las tablas con class "wikitable"
tables = soup.find_all("table", {"class": "wikitable"})

# Inspeccionamos las tablas para encontrar la correcta
for idx, table in enumerate(tables):
    print(f"Tabla {idx}:")
    print(table.find("caption"))  # Esto imprimirá el título si lo tiene


```

```python
# Suponiendo que la tabla deseada es la 0 (ajusta si es otra)
target_table = tables[1]

# Usar pandas para leer directamente la tabla HTML
df = pd.read_html(str(target_table))[0]

# Mostrar las primeras filas
df
```

```python
from io import StringIO


html_str = str(target_table)
df = pd.read_html(StringIO(html_str))[0]
```
